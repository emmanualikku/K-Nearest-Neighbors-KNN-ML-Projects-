# splitting

```python
x=df.drop(columns=['Diagnosis'])
y=df['Diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=0.2)
```

# scalling


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform it
X_train_scaled = x_train.copy()
X_train_scaled[['MMSE', 'FunctionalAssessment', 'MemoryComplaints', 'BehavioralProblems', 'ADL']] = scaler.fit_transform(
    x_train[['MMSE', 'FunctionalAssessment', 'MemoryComplaints', 'BehavioralProblems', 'ADL']])

# Transform the testing data using the same scaler
X_test_scaled = x_test.copy()
X_test_scaled[['MMSE', 'FunctionalAssessment', 'MemoryComplaints', 'BehavioralProblems', 'ADL']] = scaler.transform(
    x_test[['MMSE', 'FunctionalAssessment', 'MemoryComplaints', 'BehavioralProblems', 'ADL']])

print(X_train_scaled)
print(X_test_scaled)
```

# Model_Building

```python
# K-Nearest Neighbor(KNN) Algorithm

from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=41)
```

```python
knn.fit(X_train_scaled,y_train)
```

```python
knn.score(X_train_scaled,y_train)
```

```python
knn.score(X_test_scaled,y_test)
```

```python
y_pred=knn.predict(X_test_scaled)
```
